```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import networkx as nx
import time
import re
from functools import wraps

# ============================
# 예외 클래스 정의
# ============================

class StreamingDataAnalyzerError(Exception):
    """기본 예외 클래스"""
    pass

class DataLoadError(StreamingDataAnalyzerError):
    """데이터 로드 관련 예외"""
    pass

class DataCleanError(StreamingDataAnalyzerError):
    """데이터 정제 관련 예외"""
    pass

class VisualizationError(StreamingDataAnalyzerError):
    """시각화 관련 예외"""
    pass

class LowLevelError(StreamingDataAnalyzerError):
    """낮은 수준의 예외"""
    pass

class MediumLevelError(StreamingDataAnalyzerError):
    """중간 수준의 예외"""
    pass

class HighLevelError(StreamingDataAnalyzerError):
    """높은 수준의 예외"""
    pass

# ============================
# 데코레이터 정의
# ============================

def execution_time(func):
    """
    함수의 실행 시간을 측정하는 데코레이터.

    Parameters:
        func (function): 실행 시간을 측정할 함수.

    Returns:
        function: 래핑된 함수.
    """
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()  # 함수 실행 시작 시간 기록
        try:
            result = func(*args, **kwargs)  # 실제 함수 실행
            return result
        finally:
            elapsed_time = time.time() - start_time  # 함수 실행 종료 후 경과 시간 계산
            print(f"{func.__name__} 실행 시간: {elapsed_time:.2f} 초")  # 경과 시간 출력
    return wrapper

# ============================
# 데이터 클리너 클래스
# ============================

class DataCleaner:
    """
    데이터를 정제하는 클래스.
    """

    def __init__(self, df: pd.DataFrame):
        """
        초기화 메서드.

        Parameters:
            df (pd.DataFrame): 정제할 데이터프레임.
        """
        self._df = df  # 정제할 데이터프레임을 인스턴스 변수로 저장

    @property
    def df(self) -> pd.DataFrame:
        """
        정제된 데이터프레임을 반환하는 프로퍼티.

        Returns:
            pd.DataFrame: 정제된 데이터프레임.
        """
        return self._df

    @execution_time
    def clean_data(self):
        """
        데이터를 정제하는 메서드.
        """
        try:
            # 빈 문자열을 NaN으로 대체
            self._df.replace("", np.nan, inplace=True)
            
            # 'director' 열의 NaN 값을 'Unknown'으로 채움
            self._df['director'].fillna('Unknown', inplace=True)
            
            # 'country' 열의 NaN 값을 'Unknown'으로 채움
            self._df['country'].fillna('Unknown', inplace=True)
            
            # 'duration_int' 열을 생성하여 'duration'에서 숫자 부분 추출 후 float으로 변환
            self._df['duration_int'] = self._df['duration'].apply(self.extract_duration).astype(float)
            
            # 'duration_unit' 열을 생성하여 'duration'에서 단위 부분 추출
            self._df['duration_unit'] = self._df['duration'].apply(self.extract_duration_unit)
            
            # 'date_added' 열을 datetime 형식으로 변환
            self._df['date_added'] = pd.to_datetime(self._df['date_added'], errors='coerce')
            
            # 'added_year' 열을 생성하여 'date_added'에서 연도만 추출
            self._df['added_year'] = self._df['date_added'].dt.year
            
            print("데이터 정제 완료!")
        except Exception as e:
            # 데이터 정제 중 오류 발생 시 커스텀 예외 발생
            raise DataCleanError(f"데이터 정제 중 오류 발생: {e}")

    @staticmethod
    def extract_duration(duration: str) -> float:
        """
        duration 문자열에서 숫자 부분을 추출하는 정적 메서드.

        Parameters:
            duration (str): duration 문자열.

        Returns:
            float: 추출된 숫자.
        """
        try:
            match = re.search(r'(\d+)', duration)  # 숫자 패턴 검색
            return float(match.group(1)) if match else np.nan  # 숫자가 있으면 float으로 변환, 없으면 NaN 반환
        except Exception as e:
            # duration 추출 중 오류 발생 시 낮은 수준의 예외 발생
            raise LowLevelError(f"duration 추출 중 오류 발생: {e}")

    @staticmethod
    def extract_duration_unit(duration: str) -> str:
        """
        duration 문자열에서 단위 부분을 추출하는 정적 메서드.

        Parameters:
            duration (str): duration 문자열.

        Returns:
            str: 추출된 단위.
        """
        try:
            match = re.search(r'([a-zA-Z]+)', duration)  # 알파벳 패턴 검색
            return match.group(1) if match else np.nan  # 단위가 있으면 반환, 없으면 NaN
        except Exception as e:
            # duration 단위 추출 중 오류 발생 시 중간 수준의 예외 발생
            raise MediumLevelError(f"duration 단위 추출 중 오류 발생: {e}")

# ============================
# 데이터 비주얼라이저 클래스
# ============================

class DataVisualizer:
    """
    데이터를 시각화하는 클래스.
    """

    def __init__(self, df: pd.DataFrame):
        """
        초기화 메서드.

        Parameters:
            df (pd.DataFrame): 시각화할 데이터프레임.
        """
        self._df = df  # 시각화할 데이터프레임을 인스턴스 변수로 저장
        sns.set(style="whitegrid")  # Seaborn 스타일 설정

    @execution_time
    def visualize_data(self):
        """
        데이터를 시각화하는 메서드.
        """
        try:
            self.plot_content_type_distribution()  # 콘텐츠 유형 분포 시각화
            self.plot_yearly_release_trend()       # 연도별 출시 추이 시각화
            self.plot_top_countries_content_production()  # 상위 국가별 콘텐츠 제작 현황 시각화
            self.plot_rating_distribution()         # 평점 분포 시각화
            print("\n모든 시각화 작업이 완료되었습니다!")
        except Exception as e:
            # 시각화 중 오류 발생 시 커스텀 예외 발생
            raise VisualizationError(f"시각화 중 오류 발생: {e}")

    def plot_content_type_distribution(self):
        """
        콘텐츠 유형별 분포를 시각화하는 메서드.
        """
        try:
            plt.figure(figsize=(8, 6))  # 그래프 크기 설정
            sns.countplot(x='type', data=self._df, palette='Set2')  # 콘텐츠 유형별 카운트 플롯
            plt.title('Content Type Distribution')  # 그래프 제목
            plt.xlabel('Type')                      # x축 레이블
            plt.ylabel('Count')                     # y축 레이블
            plt.show()                               # 그래프 출력
        except Exception as e:
            # 콘텐츠 유형 분포 시각화 중 오류 발생 시 중간 수준의 예외 발생
            raise MediumLevelError(f"콘텐츠 유형 분포 시각화 중 오류 발생: {e}")

    def plot_yearly_release_trend(self):
        """
        연도별 출시 추이를 시각화하는 메서드.
        """
        try:
            release_trend = self._df['release_year'].value_counts().sort_index()  # 연도별 콘텐츠 개수 계산 및 정렬
            plt.figure(figsize=(12, 6))  # 그래프 크기 설정
            sns.lineplot(x=release_trend.index, y=release_trend.values, marker='o')  # 연도별 선 그래프
            plt.title('Yearly Release Trend')  # 그래프 제목
            plt.xlabel('Release Year')        # x축 레이블
            plt.ylabel('Number of Contents')   # y축 레이블
            plt.show()                         # 그래프 출력
        except Exception as e:
            # 연도별 출시 추이 시각화 중 오류 발생 시 중간 수준의 예외 발생
            raise MediumLevelError(f"연도별 출시 추이 시각화 중 오류 발생: {e}")

    def plot_top_countries_content_production(self):
        """
        상위 10개 국가별 콘텐츠 제작 현황을 시각화하는 메서드.
        """
        try:
            # 'country' 열을 쉼표로 분리하고 explode하여 개별 국가로 분할한 후 상위 10개 국가 추출
            top_countries = self._df['country'].str.split(', ').explode().value_counts().head(10)
            plt.figure(figsize=(10, 7))  # 그래프 크기 설정
            sns.barplot(x=top_countries.values, y=top_countries.index, palette='pastel')  # 상위 국가별 막대 그래프
            plt.title('Top 10 Countries by Content Production')  # 그래프 제목
            plt.xlabel('Number of Contents')                     # x축 레이블
            plt.ylabel('Country')                                # y축 레이블
            plt.show()                                           # 그래프 출력
        except Exception as e:
            # 상위 국가 콘텐츠 제작 현황 시각화 중 오류 발생 시 중간 수준의 예외 발생
            raise MediumLevelError(f"상위 10개 국가 콘텐츠 제작 현황 시각화 중 오류 발생: {e}")

    def plot_rating_distribution(self):
        """
        평점 분포를 시각화하는 메서드.
        """
        try:
            plt.figure(figsize=(8, 6))  # 그래프 크기 설정
            # 평점별 카운트 플롯, 평점 빈도수에 따라 정렬
            sns.countplot(x='rating', data=self._df, order=self._df['rating'].value_counts().index, palette='coolwarm')
            plt.title('Rating Distribution')  # 그래프 제목
            plt.xlabel('Rating')             # x축 레이블
            plt.ylabel('Count')              # y축 레이블
            plt.show()                        # 그래프 출력
        except Exception as e:
            # 평점 분포 시각화 중 오류 발생 시 중간 수준의 예외 발생
            raise MediumLevelError(f"평점 분포 시각화 중 오류 발생: {e}")

# ============================
# 메인 분석 클래스
# ============================

class StreamingDataAnalyzer(DataCleaner, DataVisualizer):
    """
    스트리밍 콘텐츠 데이터를 분석하고 시각화하는 통합 클래스.
    """

    def __init__(self, df: pd.DataFrame):
        """
        초기화 메서드.

        Parameters:
            df (pd.DataFrame): 분석할 데이터프레임.
        """
        DataCleaner.__init__(self, df)            # DataCleaner 초기화
        DataVisualizer.__init__(self, self.df)    # DataVisualizer 초기화 (정제된 데이터프레임 전달)

    @classmethod
    def from_dataframe(cls, df: pd.DataFrame):
        """
        클래스 메서드: 데이터프레임으로부터 인스턴스를 생성.

        Parameters:
            df (pd.DataFrame): 데이터프레임.

        Returns:
            StreamingDataAnalyzer: 인스턴스.
        """
        return cls(df=df)  # 클래스 인스턴스 반환

    @execution_time
    def run_analysis(self):
        """
        분석을 실행하는 메서드.
        """
        self.clean_data()      # 데이터 정제 실행
        self.visualize_data()  # 데이터 시각화 실행

    @property
    def dataframe(self) -> pd.DataFrame:
        """
        정제된 데이터프레임을 반환하는 프로퍼티.

        Returns:
            pd.DataFrame: 정제된 데이터프레임.
        """
        return self.df

# ============================
# 메인 함수
# ============================

def main():
    """
    스트리밍 콘텐츠 데이터 분석 및 시각화를 수행하는 메인 함수.
    """
    start_time = time.time()  # 메인 함수 실행 시작 시간 기록

    # 이미 준비된 데이터프레임을 사용합니다.
    # 인스턴스 생성: 데이터프레임을 직접 전달
    analyzer = StreamingDataAnalyzer.from_dataframe(df)

    try:
        analyzer.run_analysis()  # 분석 실행
    except DataLoadError as dle:
        print(f"데이터 로드 오류: {dle}")  # 데이터 로드 오류 처리
    except DataCleanError as dce:
        print(f"데이터 정제 오류: {dce}")  # 데이터 정제 오류 처리
    except VisualizationError as ve:
        print(f"시각화 오류: {ve}")        # 시각화 오류 처리
    except LowLevelError as lle:
        print(f"낮은 수준의 오류: {lle}")   # 낮은 수준의 오류 처리
    except MediumLevelError as mle:
        print(f"중간 수준의 오류: {mle}")   # 중간 수준의 오류 처리
    except HighLevelError as hle:
        print(f"높은 수준의 오류: {hle}")   # 높은 수준의 오류 처리
    except StreamingDataAnalyzerError as sae:
        print(f"분석 오류: {sae}")          # 분석 오류 처리
    except Exception as e:
        print(f"예상치 못한 오류 발생: {e}")  # 기타 예상치 못한 오류 처리

    print(f"main 실행 시간: {time.time() - start_time:.2f} 초")  # 메인 함수 전체 실행 시간 출력

if __name__ == "__main__":
    main()  # 메인 함수 실행
```
